```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
df = pd.read_csv("covid_19_actual.csv")
```

```python
cols = ['Location Detail','Admin2 Label (US)','Cruise Ship?', 'Case Type', 'Cases', 'Country Region', 'Date', 'Difference','Lat', 'Long','Province State']
removeCol_list = []
for each in df.columns:
    if each not in cols:
        removeCol_list.append(each)

removeCol_list
```

```python
df1 = df.drop(removeCol_list, axis=1)
```

```python
df1
```

```python
df1 = df1[df1['Country Region']=='US']
df1
```

```python
df1 = df1.rename(columns={'Location Detail':'location','Admin2 Label (US)':'county','Cruise Ship?':'ship', 'Case Type':'case_type','Cases': 'cases','Country Region':'country', 'Date':'date', 'Difference':'difference','Lat':'latitude','Long':'longitude','Province State':'state'})
df1
```

```python
df1.info()
```

```python
df1['county'] = df1['county'].map(lambda x: str(x).lstrip('| County: '))
```

```python
df1.to_csv('kbs_covid_eda.csv',header=True, index=False)
```
